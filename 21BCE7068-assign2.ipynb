```python
#histogram
z=dataset["alcohol"]
plt.hist(z,color='pink',edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x='abbrev', y='alcohol', data=dataset,color='black')
plt.xlabel('abbrev')
plt.ylabel('alcohol')
plt.show()
```

this graph shows there are more alcohol drinkers in state nd

```python
#Horizontal barchart
plt.figure(figsize=(10, 10))
x = dataset["abbrev"]
y = dataset["speeding"]
plt.barh(x,y,color = 'pink',height = 0.6, edgecolor = "black")

```

it indicates people from hl drives the fast

```python
#piechart
import matplotlib.pyplot as plt
import numpy as np
labels = dataset['abbrev']
numbers = dataset['ins_losses']
# Create a colormap with as many colors as there are states
colormap = plt.cm.get_cmap('tab20c', len(labels))
# Generate a list of colors from the colormap
colors = colormap(np.arange(len(labels)))
fig, axes1 = plt.subplots()
axes1.pie(numbers, labels=labels, autopct="%0.2f%%", startangle=90, shadow=True, colors=colors)
axes1.set_title('Distribution of Insurance Losses by the different State')
axes1.legend(labels, loc='best')
plt.show()
```

state NJ has the highest premium

```python
#boxplot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(dataset)
sns.boxplot(
    data=df,
    orient="h"
)

plt.show()


```
